```python
import pandas as pd
from path import Path
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_path = Path('Resources/heart_dummies_categorical.csv')
df_heart = pd.read_csv(file_path)
df_heart.head()
```

```python
df_heart.corr(method ='pearson')
```

```python
sns.set(style="white") 
plt.rcParams['figure.figsize'] = (15, 10) 
sns.heatmap(df_heart.corr(), annot = True, linewidths=.5, cmap="YlOrRd")
plt.title('Corelation Between Categorical Variables', fontsize = 30)
plt.show()
```
